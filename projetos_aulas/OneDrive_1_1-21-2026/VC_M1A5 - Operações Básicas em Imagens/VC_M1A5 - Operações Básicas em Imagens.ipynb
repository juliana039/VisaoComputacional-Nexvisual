## Transformações de Imagens 

Nessa prática iremos ver como realizar transformações simples mas extremamente úteis em imagens utilizando OpenCV.

```python
crop = image[0:500, 0:500]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
```

```python
# Dimensões da imagem original.
height, width = image.shape[:2]

# Matriz de translação.
translation_x, translation_y = 800, 400
translation_matrix = np.float32([
    [1, 0, translation_x],
    [0, 1, translation_y]
])

# Aplicando a matriz de transformação na imagem.
translated = cv2.warpAffine(image, translation_matrix, (width, height))

# Visualizando os resultados.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
```

```python
# Dimensões da imagem original.
height, width = image.shape[:2]

# Definindo o centro da rotação.
center = (width // 2, height // 2)
 
# Define o ângulo de rotação.
angle = 90  # Rotacionar por 45 graus.
 
# Define uma operação de escala da imagem.
scale = 0.50 # Sem escala.
 
# Deriva a matriz de rotação
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
 
# Aplica a matriz de transformação na imagem.
rotated = cv2.warpAffine(image, rotation_matrix, (width, height))

# Visualizando os resultados.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 8))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[1].imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
```

## Funções utilitárias de Desenhos

Agora iremos ver como desenhar padrões nas imagens usando OpenCV.

```python
image_draw = cv2.rectangle(image, (100,200), (500,500), (0,0,255), 5) #mais a esq e mais alto, mais a direita e mais baixo, cor, grossura da linha
plt.imshow(cv2.cvtColor(image_draw, cv2.COLOR_BGR2RGB))
```
